```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (upload first in Colab, then provide path)
file_path = "/API_SP.POP.TOTL_DS2_en_csv_v2_567709.csv"
df = pd.read_csv("/API_SP.POP.TOTL_DS2_en_csv_v2_567709.csv", skiprows=4)

# --- Histogram of populations in 2022 ---
pop_2022 = df[["Country Name", "2022"]].dropna()

plt.figure(figsize=(10,6))
plt.hist(pop_2022["2022"], bins=30, edgecolor="black")
plt.xlabel("Population (2022)")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population Across Countries (2022)")
plt.yscale("log")  # Log scale for better visualization
plt.show()

# --- Bar chart: Top 10 most populated countries in 2022 ---
top10 = pop_2022.sort_values(by="2022", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10["2022"], color="skyblue", edgecolor="black")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Country")
plt.ylabel("Population (2022)")
plt.title("Top 10 Most Populated Countries in 2022")
plt.tight_layout()
plt.show()
```
